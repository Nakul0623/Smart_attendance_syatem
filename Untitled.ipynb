```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image

# Load the trained model
model = load_model('face_recognition_model.h5')

# Load the class indices
class_indices = np.load('class_indices.npy', allow_pickle=True).item()

# Preprocess the input image
def preprocess_input_image(image_path, target_size=(128, 128)):
    img = image.load_img(image_path, target_size=target_size)
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img /= 255.0  # Normalize pixel values
    return img

# Recognize students in a single input image
def recognize_students_single(input_image_path):
    # Preprocess the input image
    input_img = preprocess_input_image(input_image_path)

    # Make predictions
    predictions = model.predict(input_img)

    # Convert predictions to class labels
    predicted_class_indices = np.argmax(predictions, axis=1)
    predicted_classes = [class_name for class_name, class_idx in class_indices.items() if class_idx == predicted_class_indices[0]]

    return predicted_classes

# Recognize students in multiple input images in a directory
def recognize_students_in_directory(directory_path):
    recognized_students = {}

    for filename in os.listdir(directory_path):
        if filename.endswith(".jpg") or filename.endswith(".jpeg") or filename.endswith(".png"):
            input_image_path = os.path.join(directory_path, filename)
            recognized_students[input_image_path] = recognize_students_single(input_image_path)

    return recognized_students

# Test the recognition function on a directory of images
input_directory_path = 'input_photos'  # Replace with the path to your input directory
recognized_students = recognize_students_in_directory(input_directory_path)

# Print the recognized students for each image
for input_image_path, students in recognized_students.items():
    print(f"Recognized Students in {input_image_path}: {students}")
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import cv2

# Load the trained model
model = load_model('face_recognition_model.h5')

# Load the class indices
class_indices = np.load('class_indices.npy', allow_pickle=True).item()

# Recognize students in a single input image
def recognize_students_single(face_img):
    # Preprocess the input image
    face_img = cv2.resize(face_img, (128, 128))
    face_img = np.expand_dims(face_img, axis=0)
    face_img = face_img / 255.0  # Normalize pixel values

    # Make predictions
    predictions = model.predict(face_img)

    # Convert predictions to class labels
    predicted_class_indices = np.argmax(predictions, axis=1)
    predicted_classes = [class_name for class_name, class_idx in class_indices.items() if class_idx == predicted_class_indices[0]]

    return predicted_classes

# Detect and recognize students in multiple input images in a directory
def recognize_students_in_directory(directory_path):
    recognized_students = {}

    # Load the face detection classifier (you may need to install OpenCV)
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    for filename in os.listdir(directory_path):
        if filename.endswith(".jpg") or filename.endswith(".jpeg") or filename.endswith(".png"):
            input_image_path = os.path.join(directory_path, filename)
            image_data = cv2.imread(input_image_path)
            gray = cv2.cvtColor(image_data, cv2.COLOR_BGR2GRAY)
            faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

            for (x, y, w, h) in faces:
                # Crop and recognize students in the detected face
                face_img = image_data[y:y + h, x:x + w]
                students = recognize_students_single(face_img)

                # Add recognized students and their names to the dictionary
                recognized_students[(x, y, x + w, y + h)] = students

    return recognized_students

# Test the recognition function on a directory of images
input_directory_path = 'input_photos'  # Replace with the path to your input directory
recognized_students = recognize_students_in_directory(input_directory_path)

# Display the recognized students along with their bounding boxes
for (x, y, x2, y2), students in recognized_students.items():
    print(f"Recognized Students: {students}")
```
